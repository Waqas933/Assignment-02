```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
print('Libraries imported successfully!')
```

## Q1. Load the dataset and display first 5 rows

```python
df = pd.read_csv('content_based_recommendation_dataset.csv')
print(df.head())
```

## Q2. Display dataset shape and column names

```python
print('Shape (Rows, Columns):', df.shape)
print('\nColumn Names:')
for col in df.columns:
    print(' -', col)
```

## Q3. Check data types of all columns

```python
print(df.dtypes)
```

## Q4. Identify missing values in each column

```python
print('Missing values per column:')
print(df.isnull().sum())
print('\nTotal missing values:', df.isnull().sum().sum())
```

## Q5. Handle missing values

```python
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        df[col] = df[col].fillna(df[col].mean())
    else:
        df[col] = df[col].fillna(df[col].mode()[0])

print('Missing values after handling:')
print(df.isnull().sum())
```

## Q6. Check and remove duplicate rows

```python
print('Duplicate rows before removal:', df.duplicated().sum())
df = df.drop_duplicates().reset_index(drop=True)
print('Duplicate rows after removal:', df.duplicated().sum())
print('New shape after removing duplicates:', df.shape)
```

## Q7. Generate summary statistics

```python
print(df.describe().to_string())
```

## Q8. Calculate the average of 'Price of the product'

```python
avg_price = df['Price of the product'].mean()
print(f'Average Price of the product: {avg_price:.2f} rupees')
```

## Q9. Find max and min of 'Median purchasing price (in rupees)'

```python
max_price = df['Median purchasing price (in rupees)'].max()
min_price = df['Median purchasing price (in rupees)'].min()
print(f'Max Median Purchasing Price: {max_price} rupees')
print(f'Min Median Purchasing Price: {min_price} rupees')
```

## Q10. Value counts of 'Brand of the product'

```python
print('Value counts of Brand of the product:')
print(df['Brand of the product'].value_counts())
```

## Q11. Sort data by 'Price of the product'

```python
df_sorted = df.sort_values(by='Price of the product', ascending=True).reset_index(drop=True)
print('Data sorted by Price of the product (ascending) - Top 10:')
print(df_sorted[['Brand of the product', 'Price of the product', 'Rating of the product']].head(10).to_string())
```

## Q12. Filter data where rating > 4

```python
df_filtered = df[df['Rating of the product'] > 4]
print(f'Total rows with Rating of the product > 4: {len(df_filtered)}')
print()
print(df_filtered[['Brand of the product', 'Rating of the product', 'Price of the product']].head().to_string())
```

## Q13. Create new column 'Price Difference'
**Formula:** Price of the product - Median Purchasing Price (in rupees)

```python
df['Price Difference'] = df['Price of the product'] - df['Median purchasing price (in rupees)']
print("New column 'Price Difference' created!")
print()
print(df[['Price of the product', 'Median purchasing price (in rupees)', 'Price Difference']].head(10).to_string())
```

## Q14. GroupBy average price by Gender

```python
gender_avg = df.groupby('Gender')['Price of the product'].mean()
print('Average Price of the product by Gender:')
print(gender_avg)
```

## Q15. Create histogram of Price of the product

```python
plt.figure(figsize=(9, 5))
plt.hist(df['Price of the product'], bins=20, color='steelblue', edgecolor='black')
plt.title('Histogram of Price of the Product', fontsize=14)
plt.xlabel('Price of the product (in rupees)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
print('Histogram displayed below:')
plt.show()
```
